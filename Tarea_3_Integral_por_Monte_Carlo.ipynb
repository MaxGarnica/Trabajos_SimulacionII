```python
import random   
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Realizaremos la siguiente integral por el método de Monte Carlo:

$\int_0^1\sqrt{ \arctan(x)}dx$

Para poder comprender mucho mejor el resultado graficaremos la función.

$\sqrt{ \arctan(x)}$

Definida en $x$ de $[0,1]$

```python
x = np.linspace(0, 1, 100)
y = np.sqrt(np.arctan(x))

fig = plt.figure(1, figsize=(8, 8))
plt.plot(x, y, color="blue", markersize=1)
plt.title("Grafica de la función")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()
```

Ahora apliquemos el método Monte Carlo. Encontraremos, graficaremos y separemos los puntos aleatorios que caeran dentro o fuera de la función.

```python
p=0
t=1500
yc=np.sqrt(np.pi/4)

x = np.linspace(0, 1, 1000)
y = np.sqrt(np.arctan(x))

fig = plt.figure(1, figsize=(8, 8))
plt.plot(x, y, color="blue", markersize=1)
plt.title("Grafica de la función aplicando el metodo Monte Carlo")
plt.xlabel("X")
plt.ylabel("Y")

for i in range(t):
  xr=random.uniform(0, 1)
  yr=random.uniform(0, yc)

  a=yr**2-np.arctan(xr)

  if a<0 or a==0:
    p=p+1 
    plt.scatter(xr, yr, s=20, c="green")
  else:
    plt.scatter(xr, yr, s=20, c="magenta")

print('El número de puntos que caen dentro de la curva son p=%d' %p)
print('El número total de puntos obntenidos es t=%d' %t)

plt.grid()
plt.show()
```

Ahora dividiendo el número de puntos dentro de la función entre el total de puntos obtendremos la aproximación del área bajo la curva es decir el valor aproximado de la integral.

```python
I=p/t

print('El valor aproximado de la integral es I=%f' %I)
```

Entonces:

$I≈\int_0^1\sqrt{ \arctan(x)}dx$
