```python
import random   
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Generaremos una distribuci√≥n normal por el m√©todo de las doce uniformes utilizando la funcion $random()$ para generar un histograma.

Utilizaremos la formula:

$x=Œº+œÉ[(ùö∫_i^{12} R_i)-6]$

con los valores de la media $Œº=180$ y desviacion estandar $œÉ=15$

$x=180+15[(ùö∫_i^{12} R_i)-6]$

```python
t=2000
mu=180
sigma=15
x=[]

for i in range(t):
  r=0
  for j in range (12):
    r_1=random.random()
    r=r+r_1

  x_1=mu+sigma*(r-6)
  x.append(x_1)

fig = plt.figure(1, figsize=(8, 8))        
plt.hist(x, color="magenta", bins=200)
plt.grid()
plt.show()
```

Ahora apliquemos la funcion $gauss()$ para comparar los histogramas.

```python
mu=180
sigma=15
xg=[]

for i in range(t):
    xg_1=random.gauss(mu,sigma)
    xg.append(xg_1)

fig = plt.figure(1, figsize=(8, 8)) 
plt.hist(xg,bins=200)
plt.show()
```

Graficando los dos histogramas en usa sola grafica.

```python
fig = plt.figure(1, figsize=(8, 8)) 
plt.hist(xg,bins=200)
plt.hist(x,color="magenta", bins=200)
plt.show()
```

Los dos histogramas son muy parecidos.
